Definition of LLMs

LLMs (Large Language Models) are AI models trained on massive amounts of text data to understand and generate human-like language.
They are used for tasks like translation, summarization, Q&A, code generation, etc.
They use:
- deep learning (particularly transformer architectures), to model relationships between words and predict the next word in a sentence.

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Loading a pretrained model and tokenizer
model_name = "gpt2"  # GPT-2 is used here for demonstration; can be replaced with models like "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

print(f"\nModel '{model_name}' loaded successfully!")
print("""
GPT-2 is a causal language model, meaning it predicts the next word in a sequence.
It has been trained on a diverse dataset and can generate coherent, contextually relevant text.
""")
```

Définition of tokenization:

Tokenization is the process of breaking down a piece of text — like a sentence or paragraph — into smaller parts called tokens, which a language model can understand and work with.

Machines do not understand raw human language: to be able to process text, a model like GPT must convert that text into numbers.
Tokenization is the first step in that process.



```python
# Create a new variable text with a sentence of your choice
text = "Tokenization is an important step in natural language processing."

# 2. Tokenize input text
tokens = tokenizer.tokenize(text)
token_ids = tokenizer.convert_tokens_to_ids(tokens)

print(f"Original Text: {text}")
print(f"Tokens: {tokens}")
print(f"Token IDs: {token_ids}")

# 3. Visualizing the tokenization process
plt.figure(figsize=(10, 4))
plt.bar(tokens, token_ids, color="skyblue")
plt.xlabel("Tokens")
plt.ylabel("Token IDs")
plt.title("Tokenization Process")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Print the ID of each token
for token, token_id in zip(tokens, token_ids):
  print(f"Token: {token}, ID: {token_id}")
```

**What does special prefixes like ‘Ġ’ indicate?**

The special prefix 'Ġ' indicates that the following characters represent a word that starts with a space.
This is common in some tokenization methods, like the Byte Pair Encoding (BPE) used by GPT-2, where spaces are treated as part of the token.
This helps the model differentiate between words that are at the beginning of a sentence or after punctuation versus words that appear within a sentence with preceding spaces.

 # Exercise 4 : Pretraining vs. Fine-Tuning

 Explain with your own words the two phases of Pretraining and FineTuning in Transformers.

**Exercise 4 : Pretraining vs. Fine-Tuning**

In the context of Transformers and large language models, **Pretraining** and **Fine-Tuning** are two distinct phases in the model's development:

*   **Pretraining:** This is the initial phase where the model is trained on a massive, diverse dataset of text (and sometimes code). The goal of pretraining is for the model to learn general language understanding, grammar, facts about the world, and common patterns. It does this through unsupervised learning tasks, such as predicting the next word in a sentence (causal language modeling) or filling in masked words (masked language modeling). During pretraining, the model develops a broad understanding of language but isn't specialized for any particular task.

*   **Fine-Tuning:** After pretraining, the model is specialized for a specific downstream task (like text classification, question answering, or translation). This is done by training the pretrained model on a smaller, task-specific dataset. During fine-tuning, the model's parameters are adjusted to perform well on the new task. The knowledge gained during pretraining provides a strong foundation, allowing the model to learn the new task effectively with less data compared to training from scratch.

Think of pretraining as learning to read and understand many different books, and fine-tuning as then using that general reading ability to become an expert in a specific subject, like medical texts or legal documents.
