Context - Email Spam Detection


Building a classifier to label emails as either:
- Spam (positive class) or
- Not Spam (negative class)

Define the confusion matrix terms

| Predicted ↓ / Actual → | Not Spam (Negative) | Spam (Positive)     |
| ---------------------- | ------------------- | ------------------- |
| **Not Spam**           | True Negative (TN)  | False Negative (FN) |
| **Spam**               | False Positive (FP) | True Positive (TP)  |


### ** Definie (in your own words):**

**True Positives (TP)**

Spam Emails that were correctly predicted as Spam by the model

**True Negatives (TN)**

Emails that were Not Spam, correctly predicted as Not Spam by the model

**False Positives (FP)**

Emails that were Not Spam, yet incorrectly correctly predicted as Spam by the model.
(→ Legitimate emails sent to the Spam folder — annoying for users!)

**False Negatives (FN)**

Spam Emails that were incorrectly predicted as Not Spam by the model
(→ Spam emails show up in your inbox — potentially dangerous if phishing/malware!)

 Calculate Accuracy, Precision, Recall, F1-Score

 Let’s assume the confusion matrix values are:

- TP = 70
- TN = 900
- FP = 30
- FN = 20

Calculation

TP = 70
TN = 900
FP = 30
FN = 20

Total = 70 + 900 + 30 + 20 = 1020

- Accuracy = (70 + 900) / 1020 = 970 / 1020 ≈ 0.951
- Precision = 70 / (70 + 30) = 70 / 100 = 0.70
- Recall = 70 / (70 + 20) = 70 / 90 ≈ 0.778
- F1-Score = 2 * (0.70 * 0.778) / (0.70 + 0.778) ≈ 0.737


Final scores

| Metric    | Score   |
| --------- | ------- |
| Accuracy  | 95.1%   |
| Precision | 70.0%   |
| Recall    | 77.8%   |
| F1-Score  | ≈ 73.7% |

Instructions Consider a medical diagnosis application where a model predicts whether patients have a certain disease.

- Explain why high recall is more important than high precision in this medical diagnosis context.
- Describe a scenario where precision becomes more important than recall.

Evaluating Trade-offs in Metrics** in the context of **medical diagnosis**, where **every prediction can have real-world consequences**.

---

1. Why is **High Recall** More Important Than High Precision in Medical Diagnosis?

Definitions in context:

* **Recall** = How many actual sick patients were **correctly identified**
* **Precision** = Of those predicted sick, how many are **actually sick**

Why prioritize **recall**?

Because **missing a sick patient** (False Negative) can lead to:

* No treatment or delayed treatment
* Worsening of condition or fatal outcomes
* Legal or ethical issues for the medical system

> 🔴 **False Negatives are dangerous.**

> Better to **flag more people** (even if some are false alarms) than to **miss someone who truly has the disease**.

Example:

Out of 100 patients:

* 10 are actually sick.
* If your model finds only 3 of them, recall = 30% — 7 sick people go untreated.
* Even if precision is 100% (you were right about all 3), the other 7 **got missed**.

---

2. When Does **Precision** Become More Important Than Recall?

Scenario:

You’re predicting **who should undergo an expensive, invasive, or risky treatment**.

* **Precision** becomes critical when:

  * Treatments carry side effects or high costs.
  * You want to avoid unnecessary emotional stress or procedures.

> **False Positives** are more costly here — wrongly telling a healthy person they are sick.

---

### Example:

* Predicting **who needs chemotherapy**.
* If your precision is low, you risk treating people who don’t need it.
* High **recall with low precision** may lead to over-treatment, resource strain, and harm.

---

###3. Why Accuracy Alone Is Misleading in **Imbalanced Datasets**

#### Medical datasets are often imbalanced:

* Example: Only 5% of patients have the disease, 95% are healthy.

#### What happens if you focus **only on accuracy**?

* A dumb model that **predicts everyone is healthy** would still be **95% accurate**.
* But it **misses all actual sick patients** — recall = 0%.

> **High accuracy doesn’t mean the model is useful** in imbalanced cases.

---

#### Key consequences of focusing only on accuracy:

| Consequence    | Description                                                       |
| -------------- | ----------------------------------------------------------------- |
| False security | Model appears effective but fails to catch rare, important cases. |
| Patient harm   | Critical illnesses go undetected (FN).                            |
| Ethical risk   | Patients trust results that may be dangerously wrong.             |

---

### Summary

| Concept         | Importance in Medical Diagnosis                                |
| --------------- | -------------------------------------------------------------- |
| **Recall ↑**    | Detect all (or most) actual sick cases — avoid false negatives |
| **Precision ↑** | Avoid unnecessary treatments — avoid false positives           |
| **Accuracy**    | Can be deceptive — especially in imbalanced data               |

> **Balance is key**, but in **medical screening**, **recall should almost always take priority.**


````python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, precision_score, recall_score, f1_score

# Simulate true labels
y_true = np.array([1]*50 + [0]*950)  # 50 have disease, 950 don't

# Simulate model predictions
# A: High Accuracy, Low Recall (misses positives)
# B: High Recall, Low Precision (many false positives)
# C: Balanced Precision & Recall
y_pred_a = np.array([0]*20 + [1]*30 + [0]*950)
y_pred_b = np.array([1]*45 + [0]*5 + [1]*200 + [0]*750)
y_pred_c = np.array([1]*40 + [0]*10 + [1]*50 + [0]*900)

models = {
    "Model A\nHigh Accuracy,\nLow Recall": y_pred_a,
    "Model B\nHigh Recall,\nLow Precision": y_pred_b,
    "Model C\nBalanced": y_pred_c
}

# Plot confusion matrices
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
for ax, (name, y_pred) in zip(axs, models.items()):
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(cm, display_labels=["No Disease", "Disease"])
    disp.plot(ax=ax, cmap='Blues', colorbar=False)
    ax.set_title(name)
    ax.grid(False)

plt.suptitle("Confusion Matrices for Medical Diagnosis Models", fontsize=16)
plt.tight_layout()
plt.show()

# Compute and plot metrics
metrics = {}
for name, y_pred in models.items():
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    metrics[name] = {
        "Precision": precision,
        "Recall": recall,
        "F1-Score": f1
    }

# Bar plot for metrics
labels = list(metrics.keys())
x = np.arange(len(labels))
width = 0.25

fig, ax = plt.subplots(figsize=(10, 6))
precision_vals = [metrics[m]["Precision"] for m in labels]
recall_vals = [metrics[m]["Recall"] for m in labels]
f1_vals = [metrics[m]["F1-Score"] for m in labels]

ax.bar(x - width, precision_vals, width, label='Precision')
ax.bar(x, recall_vals, width, label='Recall')
ax.bar(x + width, f1_vals, width, label='F1-Score')

ax.set_ylabel('Score')
ax.set_title('Model Performance Comparison')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylim(0, 1.05)
ax.legend()
ax.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
````

Three confusion matrices side by side

A bar chart comparing precision, recall, and F1-score

 in the context of a **housing price prediction task**, which is a **regression problem**.

1. K-Fold vs. Stratified K-Fold Cross-Validation

####K-Fold Cross-Validation:

* Splits the dataset into **K equal parts (folds)**.
* Trains the model on **K-1 folds**, and tests it on the remaining one.
* This process repeats K times, and the final performance is **averaged** over all runs.

**Example**: With 5-fold CV:

* Fold 1 is test, train on 2–5
* Fold 2 is test, train on 1,3–5
* ...until all folds are used once as test

---

#### Stratified K-Fold Cross-Validation:

* Also splits into K folds.
* BUT: Ensures **each fold has the same distribution of the target variable** as the full dataset.
* **Mostly used in classification** tasks, especially for imbalanced classes.

---

#### Which to use for housing price prediction?

| Type                  | Use Case                                                     |
| --------------------- | ------------------------------------------------------------ |
| **K-Fold**            | Best for regression problems like housing price prediction |
| **Stratified K-Fold** | Not necessary — no class labels to preserve in regression |

> **Choose regular K-Fold** because housing price prediction is a regression problem and doesn't involve categorical class labels.

---

#### 2. What Are Learning Curves?

#### **Learning curves** show how the model’s **performance** (error or accuracy) changes as a function of the **training set size**.

* Typically, we plot:

  * **Training error**
  * **Validation error**
* Against the number of training samples

---

#### Why are they useful?

They help diagnose:

* **Underfitting** (model too simple)
* **Overfitting** (model too complex)
* Whether more data would help

---

#### 3. Diagnosing Underfitting & Overfitting from Learning Curves

| Pattern             | What You See                                                        | Problem                                        | Solution                                                               |
| ------------------- | ------------------------------------------------------------------- | ---------------------------------------------- | ---------------------------------------------------------------------- |
| 🔴 **Underfitting** | Both training and validation errors are **high and close together** | Model is too simple                            | Try more complex model (e.g., more features, less regularization)      |
| 🟡 **Overfitting**  | Training error is **low**, but validation error is **high**         | Model is too complex, memorizing training data | Try regularization, reduce model complexity, or add more training data |
| 🟢 **Good Fit**     | Training and validation errors are **low and close**                | Model generalizes well                         | No action needed                                                     |

---

#### 🛠 How to Improve with Learning Curve Insights:

| Issue                 | Fix                                                                            |
| --------------------- | ------------------------------------------------------------------------------ |
| Underfitting          | - Add more features<br>- Use non-linear models<br>- Reduce regularization      |
| Overfitting           | - Add more data<br>- Use regularization (e.g., L2)<br>- Prune model complexity |
| Model too data-hungry | - Look for variance in validation error<br>- Use more data or augment          |

---

## Summary

| --------------------- | ---------------------------------------------------------------- |
| **K-Fold CV**         | Best for regression tasks like housing prices                    |
| **Stratified K-Fold** | Only useful for classification problems                          |
| **Learning Curves**   | Help diagnose underfitting, overfitting, and training data needs |
| **Fixing Issues**     | Adjust model complexity, regularization, or training size        |

#### 1. Why Accuracy Is Misleading

Let’s do quick math:

* Suppose you have **10,000** patients
* Only **2% = 200 patients** have the disease
* **98% = 9,800 patients** are healthy

Now imagine a model that predicts **“No disease” for everyone**

* It will be **correct for 9,800** out of 10,000 → **Accuracy = 98%**
* But it **misses all 200 sick patients**

  * **Recall = 0%**
  * **Precision = undefined or 0%**

> This model is **useless** even with 98% accuracy — it **fails to detect any actual disease cases**.

#### 2. Importance of Precision and Recall

#### Quick Definitions:

| Metric        | What It Tells You                                               |
| ------------- | --------------------------------------------------------------- |
| **Recall**    | How many actual sick patients you **correctly identified**      |
| **Precision** | Of those you predicted as sick, how many were **actually sick** |

---

#### Why They Matter:

| Metric                   | Importance in Rare Disease                                                             |
| ------------------------ | -------------------------------------------------------------------------------------- |
| **Recall (Sensitivity)** | Must be high — we don’t want to miss any real cases (False Negatives are dangerous) |
| **Precision**            | Important — too many False Positives can cause stress, unnecessary treatment, cost     |

> In most **disease detection** tasks, **recall is the top priority**, even if precision suffers slightly.

---

#### 3. Strategies to Handle Class Imbalance

#### Evaluation Strategies:

1. **Use Better Metrics**:

   * Precision, Recall, F1-Score
   * ROC-AUC, PR-AUC
   * Confusion Matrix

2. **Use Stratified Cross-Validation**:

   * Ensures minority class is represented in each fold

---

#### Modeling Strategies:

| Strategy                                                    | Purpose                                                                 |
| ----------------------------------------------------------- | ----------------------------------------------------------------------- |
| **Resampling**                                              | Balance the classes                                                     |
| → Oversampling (e.g. SMOTE)                                 | Duplicate or synthesize positive cases                                  |
| → Undersampling                                             | Remove some negative cases                                              |
| **Class Weights**                                           | Penalize misclassifying positive class more heavily                     |
| → In models like Logistic Regression, XGBoost, RandomForest |                                                                         |
| **Anomaly Detection**                                       | When positives are truly rare, reframe the task as detecting anomalies  |
| **Threshold Tuning**                                        | Manually lower the threshold for positive prediction to increase recall |

####Role of Threshold Tuning in Classification Models,
---

#### 1. What Does the Classification Threshold Do?

A **binary classifier** (like Logistic Regression, XGBoost, etc.) often outputs a **probability score** for each prediction.

* Default threshold = **0.5**

  * If `P(default) ≥ 0.5` → classify as "will default"
  * If `P(default) < 0.5` → classify as "won’t default"

You can **adjust** this threshold (e.g., to 0.7 or 0.3) depending on your risk tolerance.

---

#### 2. Impact of Changing Threshold from 0.5 → 0.7

| Metric          | Effect When Threshold ↑                                                                              |
| --------------- | ---------------------------------------------------------------------------------------------------- |
| **Recall ↓**    | You're stricter — fewer clients are flagged as likely to default. You **miss more actual defaults**. |
| **Precision ↑** | You’re more confident when you say someone will default. Fewer **false positives**.                  |

> 📉 You’ll **miss some risky clients**, but when you do flag someone, it's with higher confidence.

---

#### 3. Consequences of Setting Threshold Too High or Too Low

#### In **loan default prediction**, the cost of error is asymmetric:

---

#### **Setting Threshold Too High (e.g., 0.9)**

* You’re **too conservative** in predicting defaults
* **Few defaults get caught** → high **False Negatives**
* Risk: Giving loans to people who **will default**

  * → **Financial loss**

---

#### **Setting Threshold Too Low (e.g., 0.3)**

* You **flag too many people** as high risk
* Many **False Positives**
* Risk: Denying loans to **creditworthy clients**

  * → 📉 **Lost business opportunity** and unhappy customers

---

#### Ideal threshold = trade-off between **risk** and **revenue**

---

####  4. How ROC Curve and AUC Help

#### **ROC Curve** (Receiver Operating Characteristic):

* Plots:

  * **True Positive Rate (Recall)** on Y-axis
  * **False Positive Rate** on X-axis
* Shows performance **across all thresholds**

> Each point = model performance at a specific threshold

---

#### **AUC (Area Under Curve)**:

* AUC ranges from **0.5 (random guess)** to **1.0 (perfect model)**
* The **higher**, the better the model at distinguishing defaulters vs. non-defaulters

---

#### How it helps with threshold tuning:

* You can **visually inspect** the ROC curve to find the “elbow” point — the **best trade-off** between TPR and FPR
* Combine with **Precision-Recall curves** if data is imbalanced

---

#### You can also use metrics like:

* **Youden’s J statistic**:

  $$
  J = \text{TPR} - \text{FPR}
  $$

  → Pick threshold where this is **maximized**

* **Cost-based thresholds**: if you know financial costs of FP/FN

---

## Summary

| Topic           | Insight                                                                   |
| --------------- | ------------------------------------------------------------------------- |
| **Threshold ↑** | ↑ Precision, ↓ Recall — fewer defaults caught, more confident predictions |
| **Too High**    | Risk of approving bad loans (false negatives)                             |
| **Too Low**     | Risk of rejecting good clients (false positives)                          |
| **ROC/AUC**     | Help visualize performance across thresholds and choose the sweet spot    |

---
