```python
import pandas as pd
import matplotlib as plt
import numpy as np

random_integers = np.random.randint(1, 101, size=50)
print(random_integers)

data = [37, 40, 100, 14, 96, 61, 41, 34, 90, 24, 14, 24, 15, 72, 70, 54, 29, 3,
  33, 4, 86, 62, 12, 28, 6, 1, 91, 51, 46, 62, 74, 47, 86, 71, 83, 58,
  54, 19, 39, 34, 82, 13, 44, 60, 81, 68, 55, 71, 94, 23]

mean_val = np.mean(data)
median_val = np.median(data)
std_val = np.std(data, ddof=1)  # ddof=1 pour écart-type échantillon

print(f"Moyenne: {mean_val:.2f}")
print(f"Médiane: {median_val}")
print(f"Écart-type: {std_val:.2f}")
```

```python
# Exercice 3
import numpy as np

# Dates au format original : ['2023-01-01' '2023-01-02' '2023-01-03' '2023-01-04' '2023-01-05'] à partir du 1er janvier 2023
dates = ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']
dates = np.arange('2023-01-01', '2023-01-06', dtype='datetime64[D]')

# Convertir les dates au format YYYY/MM/DD (sous forme de chaînes de caractères)
formatted_dates = np.array([str(d).replace('-', '/') for d in dates])

print("Dates au format original :", dates)
print("Dates formatées :", formatted_dates)
```

```python
# Exercice 4
import pandas as pd
import numpy as np

# 1. DataFrame avec 10 nombres aléatoires entre 1 et 100
df = pd.DataFrame({'valeurs': np.random.randint(1, 101, size=10)})

print("=== DataFrame original ===")
print(df)

# 2. Sélection conditionnelle : valeurs supérieures à 50
print("\n=== Valeurs > 50 ===")
print(df[df['valeurs'] > 50])

# 3. Agrégation : somme et moyenne
somme = df['valeurs'].sum()
moyenne = df['valeurs'].mean()

print(f"\nSomme des valeurs : {somme}")
print(f"Moyenne des valeurs : {moyenne:.2f}")
```

Images are made up of pixels, and each pixel holds color or intensity information. NumPy arrays are well-suited to represent this pixel data because they allow matrix-style operations.
Each element is a pixel intensity value between 0 (black) and 255 (white) (for 8-bit images).

```python
# Exercice 5
# Image de niveaux de gris
import numpy as np
import matplotlib.pyplot as plt
grayscale_image = np.array([
    [  0,  50, 100, 150, 200],
    [ 10,  60, 110, 160, 210],
    [ 20,  70, 120, 170, 220],
    [ 30,  80, 130, 180, 230],
    [ 40,  90, 140, 190, 240]
], dtype=np.uint8)
plt.imshow(grayscale_image, cmap='gray')
plt.title("5x5 Grayscale Image")
plt.colorbar()
plt.show()
```

```python
# Exercice 9

import numpy as np
import pandas as pd

# Créer un tableau NumPy 4x4
array_np = np.random.randint(1, 100, size=(4, 4))
print("=== Tableau NumPy ===")
print(array_np)

# Convertir en DataFrame Pandas
df = pd.DataFrame(array_np, columns=['A', 'B', 'C', 'D'])
print("\n=== DataFrame Pandas ===")
print(df)

# Reconvertir le DataFrame en array NumPy
df = pd.DataFrame(array_np, columns=['A', 'B', 'C', 'D'])
array_np = df.to_numpy()

# Affichage
print("DataFrame :")
print(df)

print("\nArray NumPy :")
print(array_np)

```

```python
# Exercice 10

import numpy as np
import matplotlib.pyplot as plt

data = np.array([73, 18, 94, 65, 29, 55, 39, 84, 13, 43])

# Axe X : indices (0 à 9)
x = np.arange(len(data))

# Tracer le graphe
plt.plot(x, data, marker='o', linestyle='-', color='b')

# Ajout de titres et labels
plt.title('Graphique en ligne des 10 valeurs')
plt.xlabel('Index')
plt.ylabel('Valeur')

# Afficher la grille et le graphique
plt.grid(True)
plt.tight_layout()
plt.show()
```
