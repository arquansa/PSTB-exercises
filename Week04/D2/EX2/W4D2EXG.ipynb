##Exercise 1 : Advanced Matrix Operations

### Create a 5x5 matrix with random numbers and perform the following operations:
1. Calculate the eigenvalues and eigenvectors.
2. Perform matrix normalization.

```python
import numpy as np

# Create a 5x5 matrix with random numbers
matrix = np.random.rand(5, 5)
print("Original 5x5 Matrix:")
print(matrix)
```

```python
# Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(matrix)

print("\nEigenvalues:")
print(eigenvalues)
print("\nEigenvectors:")
print(eigenvectors)
```

```python
# Perform matrix normalization (Min-Max Scaling)
min_val = np.min(matrix)
max_val = np.max(matrix)
normalized_matrix = (matrix - min_val) / (max_val - min_val)

print("\nNormalized Matrix:")
print(normalized_matrix)
```

## Exercise 2 : Statistical Distribution Analysis

Exercise 2 : Statistical Distribution Analysis
Instructions Hint:

- To generate a dataset following a normal distribution, you can use numpy‚Äôs np.random.normal() function.
- For plotting its histogram, matplotlib‚Äôs plt.hist() can be used.
- Generate a dataset following a normal distribution and plot its histogram. Then,
- calculate and interpret its skewness and kurtosis.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a dataset following a normal distribution
mean = 0
std_dev = 1
num_samples = 1000
normal_data = np.random.normal(mean, std_dev, num_samples)

# Plot its histogram
plt.hist(normal_data, bins=30, density=True, alpha=0.6, color='g')
plt.title('Histogram of a Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
from scipy.stats import skew, kurtosis

# Calculate skewness and kurtosis
skewness = skew(normal_data)
kurt = kurtosis(normal_data)

print(f"\nSkewness: {skewness:.4f}")
print(f"Kurtosis: {kurt:.4f}")

# Interpretation
print("\nInterpretation:")
if abs(skewness) < 0.5:
    print("The skewness is close to 0, suggesting the distribution is approximately symmetrical.")
elif skewness > 0:
    print("The skewness is positive, indicating the distribution is skewed to the right.")
else:
    print("The skewness is negative, indicating the distribution is skewed to the left.")

if kurt < 0:
    print("The kurtosis is negative (platykurtic), meaning the distribution has lighter tails and is flatter than a normal distribution.")
elif kurt > 0:
    print("The kurtosis is positive (leptokurtic), meaning the distribution has heavier tails and is more peaked than a normal distribution.")
else:
     print("The kurtosis is close to 0 (mesokurtic), meaning the distribution has a similar peakedness to a normal distribution.")
```

## Exercise 3 : Time Series Forecasting

import numpy as np

- Generating monthly sales data as a NumPy array
monthly_sales_data = np.random.randint(100, 500, size=12) # Random sales data for 12 months

- Using a simple NumPy array representing time series data (e.g., monthly sales data), demonstrate how to forecast future data points using linear regression.

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Reshape the data for the linear regression model
# The independent variable (X) should be the month number
months = np.arange(1, len(monthly_sales_data) + 1).reshape(-1, 1)
# The dependent variable (y) is the monthly sales data
sales = monthly_sales_data

# Create and train the linear regression model
model = LinearRegression()
model.fit(months, sales)

# Forecast the next few data points (e.g., next 3 months)
future_months = np.arange(len(monthly_sales_data) + 1, len(monthly_sales_data) + 4).reshape(-1, 1)
forecasted_sales = model.predict(future_months)

print("Monthly Sales Data:")
print(monthly_sales_data)
print("\nForecasted Sales for the next 3 months:")
print(forecasted_sales)

# Plot the original data and the forecast
plt.scatter(months, sales, color='blue', label='Original Data')
plt.plot(future_months, forecasted_sales, color='red', label='Forecast')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Data and Forecast using Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

- Using a Pandas DataFrame, demonstrate grouping and aggregation techniques to summarize the data effectively.

```python
import pandas as pd

# Creating a sample Pandas DataFrame
data = {
    'Product': ['Product A', 'Product B', 'Product C', 'Product A', 'Product B'],
    'Sales': [200, 150, 300, 250, 180],
    'Month': ['January', 'February', 'March', 'April', 'May']
}
df = pd.DataFrame(data)

print("Original DataFrame:")
display(df)

# Demonstrate grouping and aggregation
# Group by 'Product' and calculate the sum of 'Sales' for each product
product_sales = df.groupby('Product')['Sales'].sum().reset_index()
print("\nTotal Sales per Product:")
display(product_sales)

# Group by 'Month' and calculate the average 'Sales' for each month
month_avg_sales = df.groupby('Month')['Sales'].mean().reset_index()
print("\nAverage Sales per Month:")
display(month_avg_sales)
```

Create a multi-plot layout using Matplotlib and plot different types of graphs (e.g., line, scatter, bar) with NumPy-generated data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
x = np.arange(0, 10, 0.1)
y_line = np.sin(x)
y_scatter = np.random.rand(50)
x_bar = ['A', 'B', 'C', 'D', 'E']
y_bar = np.random.randint(1, 10, size=5)

# Create a multi-plot layout (2x2 grid)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot a line graph
axes[0, 0].plot(x, y_line, color='blue')
axes[0, 0].set_title('Line Plot')
axes[0, 0].set_xlabel('X-axis')
axes[0, 0].set_ylabel('Y-axis')

# Plot a scatter plot
axes[0, 1].scatter(np.arange(len(y_scatter)), y_scatter, color='red')
axes[0, 1].set_title('Scatter Plot')
axes[0, 1].set_xlabel('Index')
axes[0, 1].set_ylabel('Value')

# Plot a bar chart
axes[1, 0].bar(x_bar, y_bar, color='green')
axes[1, 0].set_title('Bar Chart')
axes[1, 0].set_xlabel('Category')
axes[1, 0].set_ylabel('Value')

# Plot a histogram (using the normal_data from Exercise 2 as an example)
if 'normal_data' in locals():
    axes[1, 1].hist(normal_data, bins=30, density=True, alpha=0.6, color='purple')
    axes[1, 1].set_title('Histogram (Normal Distribution)')
    axes[1, 1].set_xlabel('Value')
    axes[1, 1].set_ylabel('Frequency')
else:
    # Generate some random data for histogram if normal_data is not available
    random_hist_data = np.random.randn(1000)
    axes[1, 1].hist(random_hist_data, bins=30, density=True, alpha=0.6, color='purple')
    axes[1, 1].set_title('Histogram (Random Data)')
    axes[1, 1].set_xlabel('Value')
    axes[1, 1].set_ylabel('Frequency')


# Adjust layout to prevent overlapping titles and labels
plt.tight_layout()

# Show the plots
plt.show()
```
