```python
import numpy as np
import pandas as pd

np.random.seed(42)

temperature_data = np.random.uniform(low=-5, high=35, size=(10, 12))


cities = [f'City_{i+1}' for i in range(10)]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df_temperatures = pd.DataFrame(data=temperature_data, index=cities, columns=months)

df_temperatures

annual_avg_temp = df_temperatures.mean(axis=1)

city_highest_avg = annual_avg_temp.idxmax()
city_lowest_avg = annual_avg_temp.idxmin()

print("Annual Average Temperatures:\n", annual_avg_temp)
print("\nCity with Highest Average Temperature:", city_highest_avg)
print("City with Lowest Average Temperature:", city_lowest_avg)

#Line plot: monthly temperatures for all cities

#Bar plot: annual average temperatures

#Heatmap: temperature data

import matplotlib.pyplot as plt
import seaborn as sns

# Line plot of monthly temperatures
plt.figure(figsize=(12, 6))
for city in cities:
    plt.plot(months, df_temperatures.loc[city], label=city)
plt.title('Monthly Temperatures by City')
plt.xlabel('Month')
plt.ylabel('Temperature (C)')
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.show()

# Bar plot of annual average temperatures
plt.figure(figsize=(10, 6))
sns.barplot(x=annual_avg_temp.index, y=annual_avg_temp.values, palette='coolwarm')
plt.title('Annual Average Temperature by City')
plt.ylabel('Average Temperature (C)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Heatmap of temperatures
plt.figure(figsize=(12, 6))
sns.heatmap(df_temperatures, annot=True, cmap='coolwarm', fmt=".1f")
plt.title('Heatmap of Monthly Temperatures by City')
plt.xlabel('Month')
plt.ylabel('City')
plt.tight_layout()
plt.show()

```
