Exercise 1: Multivariate Normal Distribution
Task:

Generate and visualize both a univariate and a multivariate normal distribution using SciPy.
Explain their differences.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, multivariate_normal

# Univariate Normal Distribution
univariate_data = norm.rvs(size=1000)

plt.figure(figsize=(10, 4))
plt.hist(univariate_data, bins=30, density=True, alpha=0.6, color='skyblue')
plt.title('Univariate Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Multivariate Normal Distribution
mean = [0, 0]
cov = [[1, 0.8], [0.8, 1]]  # introducing correlation
multivariate_data = multivariate_normal.rvs(mean=mean, cov=cov, size=1000)
x, y = multivariate_data.T

plt.figure(figsize=(6, 6))
plt.scatter(x, y, alpha=0.5)
plt.title('Multivariate Normal Distribution (2D)')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.axis('equal')
plt.show()
```

Explanation:

Univariate Normal Distribution:

One variable.

Symmetric bell-shaped curve centered around the mean (Gauss curve).

Described by mean and standard deviation.

Multivariate Normal Distribution:

Two or more variables.

Includes covariance structure (how variables co-vary).

When variables are correlated, the data forms an elliptical shape.

Described by a mean vector and a covariance matrix.

Key Insight:

Multivariate distributions capture relationships between variables, whereas univariate ones describe single-variable variation.

üìà Exercise 3: Statistical Hypothesis

- Testing (ANOVA)
Task:

Perform an ANOVA test to compare the sales data across three regions.

```python
import numpy as np
import pandas as pd
from scipy.stats import f_oneway

# Seed and data
np.random.seed(0)
region1 = np.random.normal(20000, 3000, 30)
region2 = np.random.normal(22000, 3500, 30)
region3 = np.random.normal(25000, 5000, 30)

# ANOVA test
f_stat, p_value = f_oneway(region1, region2, region3)

print(f"F-statistic: {f_stat:.2f}")
print(f"P-value: {p_value:.4f}")
```

Output Example:

```python
F-statistic: 10.85
P-value: 0.0001
```

Interpretation:
- Null Hypothesis (H‚ÇÄ): All regions have the same average sales.
- Alternative Hypothesis (H‚ÇÅ): At least one region's average sales differ.
- P-value < 0.05 ‚Üí We reject the null hypothesis.

There is significant difference in mean sales between regions.

Exercise 4: Linear Regression Analysis

Task:

Perform a simple linear regression to study the relationship between study hours and test scores.

```python
import numpy as np
import pandas as pd
from scipy.stats import linregress
import matplotlib.pyplot as plt

# Seed and data
np.random.seed(0)
X = np.random.rand(100) * 50  # Hours studied
Y = 2.5 * X + np.random.randn(100) * 10  # Test scores

# Regression
slope, intercept, r_value, p_value, std_err = linregress(X, Y)

print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"R-squared: {r_value**2:.3f}")
print(f"P-value: {p_value:.4f}")

# Plotting
plt.figure(figsize=(8, 5))
plt.scatter(X, Y, alpha=0.6, label='Data Points')
plt.plot(X, slope * X + intercept, color='red', label='Regression Line')
plt.xlabel('Hours Studied')
plt.ylabel('Test Score')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

Output Example:

```python
Slope: 2.47
Intercept: 0.77
R_squared: 0.901
P_value: 0.0000
```

Interpretation:

Slope (2.47): For each additional hour studied, the test score increases by ~2.47 points.

Intercept (0.77): Baseline score when no hours are studied.

R-squared (0.901): 90.1% of the variability in test scores is explained by hours studied.

P-value < 0.05: The relationship is statistically significant.
