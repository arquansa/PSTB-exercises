In an artificial neural network, information flows from the input layer (features) through hidden layers where weighted sums are computed.

Each neuron applies an activation function (like ReLU or sigmoid) to introduce non-linearity.

The output layer provides final predictions. Weights and biases are adjusted during training to minimize error between predictions and actual values.

#Exercise 3:: Create and Visualize a Noisy Dataset

Objective:
Generate a dataset with noise and visualize it.
Instructions:

- Import the necessary libraries: numpy, matplotlib.pyplot, and mean_squared_error from sklearn.metrics.
- Create a set of 20 points using the formula ( y = -x^2 ), adding normally distributed noise with a mean of 0 and a standard deviation of 0.05 using np.random.normal.

Hint:

np.random.seed(0) x = np.arange(-1, 1, 0.1) y = -x**2 + np.random.normal(0, 0.05, len(x))

- Plot the generated points using a scatter plot to visualize the dataset with noise.
- Split the dataset into a training set (first 12 points) and a test set (last 8 points).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Step 1: Generate dataset
np.random.seed(0)
x = np.arange(-1, 1, 0.1)
y = -x**2 + np.random.normal(0, 0.05, len(x))

# Step 2: Plot
plt.scatter(x, y, color='blue', label='Noisy data')
plt.title('Noisy Quadratic Dataset')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

# Step 3: Split data
x_train, y_train = x[:12], y[:12]
x_test, y_test = x[12:], y[12:]
```

#Exercise 4: Fitting Polynomial Models of Different Degrees

Objective: Fit polynomial models of varying degrees and observe overfitting.
Instructions:

- Define a function polynomial_fit that takes a degree as input and returns a polynomial that best fits the training data.
- Define another function plot_polyfit that plots the training set, test set, and the fitted polynomial curve for a given degree.
- Use plot_polyfit to visualize the fit for polynomials of degree 1, 7, and 11.
- Observe how the model behaves on the training and test sets as the degree of the polynomial increases.

### Exercise 4: Fit Polynomials and Observe Overfitting

```python
def polynomial_fit(x, y, degree):
    return np.poly1d(np.polyfit(x, y, degree))

def plot_polyfit(degree):
    model = polynomial_fit(x_train, y_train, degree)
    x_line = np.linspace(-1, 1, 100)
    y_line = model(x_line)

    plt.scatter(x_train, y_train, label='Train Data')
    plt.scatter(x_test, y_test, label='Test Data')
    plt.plot(x_line, y_line, color='red', label=f'Degree {degree} Fit')
    plt.title(f'Polynomial Fit (Degree {degree})')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot degrees 1, 7, and 11
plot_polyfit(1)
plot_polyfit(7)
plot_polyfit(11)
```

Observation:
- Degree 1 underfits (misses curvature)
- Degree 7 fits well
- Degree 11 overfits (wiggly curve)

 # Exercise 5: Cross-Validation to Find the Optimal Degree

Objective: Use cross-validation to determine the best polynomial degree. Instructions:

- Loop through polynomial degrees from 1 to 11.
- For each degree, calculate the Root Mean Squared Error (RMSE) for both the training and test sets.
- Store the results (degree, RMSE for training, RMSE for test) in a list.
- Plot the RMSE values for both the training and test sets against the polynomial degree using a logarithmic scale for the y-axis.
- Identify the degree that minimizes the RMSE on the test set and confirm that it corresponds to the true underlying model (( y = -x^2 )).

```python
results = []

for degree in range(1, 12):
    model = polynomial_fit(x_train, y_train, degree)
    y_train_pred = model(x_train)
    y_test_pred = model(x_test)

    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))

    results.append((degree, rmse_train, rmse_test))

# Plot RMSE
degrees = [r[0] for r in results]
train_rmse = [r[1] for r in results]
test_rmse = [r[2] for r in results]

plt.figure(figsize=(8, 5))
plt.plot(degrees, train_rmse, label='Training RMSE', marker='o')
plt.plot(degrees, test_rmse, label='Test RMSE', marker='o')
plt.yscale('log')
plt.xlabel('Polynomial Degree')
plt.ylabel('RMSE (log scale)')
plt.title('Cross-Validation: RMSE vs Degree')
plt.legend()
plt.grid(True)
plt.show()

# Print best degree
best_degree = degrees[np.argmin(test_rmse)]
print(f"Best polynomial degree based on test RMSE: {best_degree}")
```

**Conclusion**:


The degree with lowest test RMSE should be close to 2, matching the true underlying model (y = -x^2).

Overfitting increases test RMSE despite low training error.
