# Exercise 1: Bar Graph
Objective: Create a bar graph representing the sales data of a retail company.
Tasks:
Use a fictional dataset where the x-axis represents product categories and the y-axis represents total sales amounts.
Customize the bar graph with appropriate labels for the axes, a title, and color.
Display the graph using Matplotlib.

- Create a fictional dataset with product categories and total sales.
- Plot a bar graph using matplotlib.pyplot.
- Customize the graph:
- Add a title
- Label x and y axes
- Use custom colors

```python
import matplotlib.pyplot as plt

# Fictional sales dataset
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Books', 'Toys']
sales = [85000, 56000, 72000, 31000, 45000]

# Create the bar graph
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, sales, color='skyblue', edgecolor='black')

# Customize the plot
plt.title('Total Sales by Product Category (2025)', fontsize=16)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)
plt.xticks(rotation=45)

# Add value labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1000, f"${yval:,}", ha='center', va='bottom', fontsize=10)

# Display the graph
plt.tight_layout()
plt.show()
```

As expected, we have a graph with:

- Labeled categories
- Sales values above each bar
- Bars styled with color and edge outlines.

# Exercise 2: Hierarchical Indexing

Objective: Filter and retrieve all days when the temperature exceeded 30¬∞C (86¬∞F) for cities in Canada using hierarchical indexing.
 - Create a pandas DataFrame with hierarchical indexing on ‚ÄòCountry‚Äô, ‚ÄòCity‚Äô, and ‚ÄòDate‚Äô.
 - Write Python code to filter the dataset for the specified condition.
 - Explain how hierarchical indexing simplifies multi-level data analysis.

Hierarchical Indexing ‚Äì Temperature Filtering
- Filter and retrieve all days when the temperature exceeded 30¬∞C (86¬∞F) for cities in Canada using hierarchical indexing in a pandas DataFrame.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a fictional dataset
data = {
    ('Canada', 'Toronto', '2025-08-01'): 31,
    ('Canada', 'Toronto', '2025-08-02'): 29,
    ('Canada', 'Vancouver', '2025-08-01'): 27,
    ('Canada', 'Vancouver', '2025-08-02'): 32,
    ('USA', 'New York', '2025-08-01'): 35,
    ('USA', 'Chicago', '2025-08-01'): 33,
    ('Canada', 'Montreal', '2025-08-01'): 30.5,
    ('Canada', 'Montreal', '2025-08-02'): 28
}

# Convert the dictionary to a Series
temp_series = pd.Series(data, name='Temperature (¬∞C)')

# Convert to DataFrame and set hierarchical index
df = temp_series.reset_index()
df.columns = ['Country', 'City', 'Date', 'Temperature (¬∞C)']
df['Date'] = pd.to_datetime(df['Date'])
df.set_index(['Country', 'City', 'Date'], inplace=True)

# Filter: Canada AND Temperature > 30¬∞C
filtered = df.loc['Canada']
hot_days = filtered[filtered['Temperature (¬∞C)'] > 30]

print("üå°Ô∏è Days above 30¬∞C in Canadian cities:\n")
print(hot_days)
```

Plot a Bar Graph
Graph of Canadian City Temperatures > 30¬∞C

Plotting a bar chart where:
- X-axis = City + Date
- Y-axis = Temperature (¬∞C)

```python
# Prepare labels for the X-axis (City + Date)
hot_days = hot_days.reset_index()
hot_days['Label'] = hot_days['City'] + ' - ' + hot_days['Date'].dt.strftime('%b %d')

# Plot the bar graph
plt.figure(figsize=(10, 6))
bars = plt.bar(hot_days['Label'], hot_days['Temperature (¬∞C)'], color='tomato', edgecolor='black')

# Add titles and labels
plt.title('üî• Canadian Cities Over 30¬∞C in August 2025', fontsize=14)
plt.xlabel('City and Date', fontsize=12)
plt.ylabel('Temperature (¬∞C)', fontsize=12)
plt.xticks(rotation=45)

# Add value labels above bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.3, f"{height:.1f}¬∞C", ha='center', va='bottom')

plt.tight_layout()
```

#Exercise 4: Visualizing Movie Durations
Objective:

Create a distribution plot to visualize the durations of MCU movies.
Tasks:
- Import necessary libraries (pandas, Seaborn, Matplotlib).
- Load a dataset containing MCU movie information.
- Use Seaborn‚Äôs distplot or histplot to visualize the distribution of movie durations.
- Add a KDE overlay and customize the plot with a title.
- Display the plot.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create fictional MCU movie dataset
data = {
    'Title': [
        'Iron Man', 'The Incredible Hulk', 'Thor', 'Captain America: The First Avenger',
        'The Avengers', 'Iron Man 3', 'Thor: The Dark World', 'Captain America: The Winter Soldier',
        'Guardians of the Galaxy', 'Avengers: Age of Ultron', 'Doctor Strange', 'Black Panther',
        'Avengers: Infinity War', 'Avengers: Endgame'
    ],
    'Duration': [126, 112, 115, 124, 143, 130, 112, 136, 121, 141, 115, 134, 149, 181]  # in minutes
}

df = pd.DataFrame(data)

# Step 2: Create distribution plot using Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(df['Duration'], kde=True, color='skyblue', bins=8, edgecolor='black')

# Step 3: Customize plot
plt.title('‚è±Ô∏è Distribution of MCU Movie Durations', fontsize=16)
plt.xlabel('Duration (minutes)', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)

# Step 4: Display the plot
plt.tight_layout()
plt.show()
```

#Exercise 5: Box Plot of Audience and Tomato Meter Scores
Objective: Compare the distribution of Tomato Meter scores and Audience scores using a box plot.
Tasks:
- Create a new DataFrame that melts the Tomato Meter and Audience scores into a long format.
- Use Seaborn‚Äôs boxplot to visualize the distributions of both scores.
- Customize the plot by setting an appropriate color palette and adding a title.
- Display the plot.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create a fictional dataset of MCU movies with scores
data = {
    'Title': [
        'Iron Man', 'Thor', 'Captain America: Civil War',
        'Avengers: Endgame', 'Doctor Strange', 'Black Panther',
        'Eternals', 'Ant-Man', 'Guardians of the Galaxy Vol. 2'
    ],
    'Tomato Meter': [94, 77, 91, 94, 89, 96, 47, 83, 85],
    'Audience Score': [91, 75, 89, 90, 86, 79, 78, 85, 87]
}

df = pd.DataFrame(data)

# Step 2: Melt the DataFrame to long format
df_melted = df.melt(id_vars='Title', value_vars=['Tomato Meter', 'Audience Score'],
                    var_name='Score Type', value_name='Score')

# Step 3: Create the boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_melted, x='Score Type', y='Score', palette='pastel')

# Step 4: Customize the plot
plt.title('üé¨ Comparison of Tomato Meter and Audience Scores (MCU)', fontsize=14)
plt.ylabel('Score (%)')
plt.xlabel('Score Type')

# Step 5: Display the plot
plt.tight_layout()
plt.show()
```

#Exercise 6: Pair Plot of Financial Data

Objective: Use Seaborn‚Äôs pairplot to explore relationships between financial aspects of MCU movies.

Tasks:
- Select relevant columns for analysis (e.g., Production Budget, Opening Weekend, Domestic Box Office, Worldwide Box Office).
- Convert financial data from string to numeric if needed.
- Use Seaborn‚Äôs pairplot to visualize pairwise relationships between the selected metrics.
- Set hue based on the MCU phase (e.g., mcu_phase) to identify patterns related to movie phases. Customize the plot and display it.

Recap:

Use Seaborn‚Äôs pairplot to explore relationships between:

- Production Budget
- Opening Weekend
- Domestic Box Office
- Worldwide Box Office
- With hue='mcu_phase' to compare MCU phases.

Sample

- Financial data in USD (as strings initially)
- A mcu_phase column

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create fictional MCU financial data
data = {
    'Title': [
        'Iron Man', 'Thor', 'Black Panther', 'Doctor Strange', 'Avengers: Endgame',
        'Eternals', 'Ant-Man', 'Guardians of the Galaxy Vol. 2'
    ],
    'Production Budget': ['$140,000,000', '$150,000,000', '$200,000,000', '$165,000,000',
                          '$356,000,000', '$200,000,000', '$130,000,000', '$200,000,000'],
    'Opening Weekend': ['$98,618,668', '$65,723,338', '$202,003,951', '$85,058,311',
                        '$357,115,007', '$71,297,219', '$57,225,526', '$145,049,000'],
    'Domestic Box Office': ['$318,412,101', '$181,030,624', '$700,426,566', '$232,641,920',
                            '$858,373,000', '$164,870,234', '$180,202,163', '$389,813,101'],
    'Worldwide Box Office': ['$585,796,247', '$449,326,618', '$1,346,913,161', '$677,718,395',
                             '$2,798,000,000', '$402,064,899', '$519,311,965', '$863,756,051'],
    'mcu_phase': ['Phase 1', 'Phase 1', 'Phase 3', 'Phase 3', 'Phase 3', 'Phase 4', 'Phase 2', 'Phase 3']
}

df = pd.DataFrame(data)

# Step 2: Clean the financial columns: remove $, commas and convert to integers
financial_cols = ['Production Budget', 'Opening Weekend', 'Domestic Box Office', 'Worldwide Box Office']

for col in financial_cols:
    df[col] = df[col].replace('[\$,]', '', regex=True).astype(int)

# Step 3: Create a pairplot of the financial columns
sns.set(style='whitegrid')
plt.figure(figsize=(10, 8))  # Though pairplot manages size itself, this is helpful for notebooks

pairplot = sns.pairplot(df,
                        vars=financial_cols,
                        hue='mcu_phase',
                        palette='Set2',
                        corner=True)  # corner=True hides duplicate upper triangle

# Step 4: Customize the overall title (Seaborn doesn't support it directly in pairplot)
pairplot.fig.suptitle("üí∞ MCU Financial Data ‚Äì Pairwise Relationships by Phase",
                      y=1.02, fontsize=16)

# Step 5: Display the plot
plt.show()
```
