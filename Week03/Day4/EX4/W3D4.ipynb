```python
#Exercice 4
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame({'CGPA': [6.5, 7.0, 8.1, 9.0, 7.8, 8.5, 6.9, 9.2, 8.0, 7.2, 7.5, 8.8, 6.3, 9.5]})
sns.histplot(data=df, x='CGPA', bins=30) # Histogramme
plt.xlabel('CGPA')
plt.ylabel('Number of Students')
plt.title("Distribution of Students' CGPA'")

plt.show()
```
