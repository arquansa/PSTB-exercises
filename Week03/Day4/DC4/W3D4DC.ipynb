Visualization

Employ a Matplotlib scatter plot to investigate the relationship between ‘Social support’ and ‘Score’. Regional Comparisons of GDP and Health:

```python
#**Scatter Plot: Social Support vs Score**

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df_2019['Social support'], df_2019['Score'], alpha=0.7, edgecolors='k')

plt.title('Social Support vs Score (2019)', fontsize=14)
plt.xlabel('Social Support')
plt.ylabel('Happiness Score')
plt.grid(True)
plt.tight_layout()
plt.show()


```

Regional Comparisons of GDP and Health:

Create a Matplotlib subplot that compares ‘GDP per Capita’ and ‘Healthy Life Expectancy’ across different regions. Use a bar plot for ‘GDP per Capita’ and a line plot for ‘Healthy Life Expectancy’ on the same axes to observe how economic strength relates to health outcomes in different regions.

```python
#Bar plot + Line plot

import pandas as pd
import matplotlib.pyplot as plt

# Load data
df_2019 = pd.read_csv('2019.csv')

# Sort for better readability (optional)
df_2019 = df_2019.sort_values('GDP per capita', ascending=False)

# Set figure size
plt.figure(figsize=(16, 8))

# X-axis values
countries = df_2019['Country or region']
x = range(len(countries))

# Bar plot: GDP per capita
plt.bar(x, df_2019['GDP per capita'], color='skyblue', label='GDP per Capita')

# Line plot: Healthy life expectancy
plt.plot(x, df_2019['Healthy life expectancy'], color='green', marker='o', label='Healthy Life Expectancy')

# Labels and title
plt.xlabel('Country or Region')
plt.ylabel('Value')
plt.title('GDP per Capita (Bar) and Healthy Life Expectancy (Line) by Country (2019)')

# X-ticks (rotate for readability)
plt.xticks(x, countries, rotation=90)

# Add legend
plt.legend()

# Adjust layout
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)


```

Here above displayed a bar plot for ‘GDP per Capita’ and a line plot for ‘Healthy Life Expectancy’ on the same axes.
They show how economic strength relates to health outcomes in different regions.
